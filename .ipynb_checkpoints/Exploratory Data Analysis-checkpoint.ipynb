```python
def plot_item_weight(df_train,item_type):
    df_train = df_train[df_train.Item_Weight.notnull()]
    data = df_train[df_train.Item_Type == item_type]['Item_Weight']
    sns.distplot(data)
    return data
```

```python
data = plot_item_weight(df_train,'Breads')
plt.show()
print data.describe()
```
